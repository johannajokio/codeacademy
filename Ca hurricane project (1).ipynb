```python
# write your update damages function here:

def update_damages(damages):
  conversion = {"M": 1000000,
              "B": 1000000000}
  updated_damages = []
  for cost in damages:
    if cost == "Damages not recorded":
      updated_damages.append(cost) 
    elif cost[-1] == "M"  :
      updated_damages.append(float(cost.strip("M")) * (conversion["M"]))
    elif cost[-1] == "B":
      updated_damages.append(float(cost.strip("B")) * (conversion["B"]))   
  return updated_damages

updated_damages = update_damages(damages)
pprint.pprint(updated_damages)

#i almost wrote this myself
```

```python

# write your construct hurricane dictionary function here:

def construct_dictionary(names, months, years,max_sustained_winds, areas_affected, updated_damages, deaths):
  hurricanes = { }
  for index in range(len(names)): 
    hurricanes[names[index]] = {"Name": names[index], 
    "Month": months[index], 
    "Year": years[index], 
     "Max Sustained Wind": max_sustained_winds[index],
     "Areas Affected": areas_affected[index],
     "Damage": updated_damages[index],
     "Deaths": deaths[index]}
  return hurricanes
     
hurricanes_data = construct_dictionary(names, months, years,max_sustained_winds, areas_affected, updated_damages, deaths)
pprint.pprint(hurricanes_data)


##ok i understand this
```
